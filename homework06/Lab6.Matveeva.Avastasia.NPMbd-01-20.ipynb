```python
class my_PCA(): 
    def __init__(self, n_components=None): 
        self.n_components = n_components
     
    def fit(self, X): 
        Y = X.astype('float64')
        Y_mean = np.mean(Y, axis = 0)
        Y -= Y_mean 
        C = Y.T.dot(Y)
        w, v = np.linalg.eigh(C)
        w = w[::-1]
        v = v[:, ::-1]
        return Y, v, w
     
    def transform(self, X, v): 
        Y = X.dot(v)
        return Y
     
    def fit_transform(self, X):
        X_c, v, w = self.fit(X)
        Y = self.transform(X_c, v)
        if self.n_components != None:
            Y = Y[:, :(self.n_components)]
        return Y
```

```python
first = my_PCA(2)
Y, v, w = first.fit(train[0].reshape(-1,784))
plt.subplot(1,2,1)
plt.plot(w, marker = '.', linestyle = '')
plt.subplot(1,2,2)
plt.plot(np.cumsum(w) / np.sum(w))
```
